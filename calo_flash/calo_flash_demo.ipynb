```python
from calo_flash import Shower
import numpy as np
import matplotlib.pyplot as plt
```

```python
S = Shower(E=1000, material='lead')
```

```python
samples = 1000

t = np.linspace(0, 20, 100)
fig, ax = plt.subplots()
for i in range(samples):
    S.fluctuate_profile()
    ax.plot(t, S.fluc_longitudinal_profile(t), color='grey', alpha=0.1,
            label='fluctuated' if i == 0 else None)
ax.plot(t, S.mean_longitudinal_profile(t), label='mean', color='red')
ax.legend()
ax.set_xlabel('t')
ax.set_ylabel('dE/dt')
ax.set_title('Longitudinal Profile')
plt.show()
```

```python
r = np.linspace(0, 0.5, 100)
fig, ax = plt.subplots()
for i in range(samples):
    S.fluctuate_profile()
    ax.plot(r, S.fluc_radial_profile(r, t=1), color='grey', alpha=0.1,
            label='fluctuated' if i == 0 else None)
ax.plot(r, S.mean_radial_profile(r, t=1), label='mean', color='red')
ax.legend()
ax.set_xlabel('r')
ax.set_ylabel('dE/dr')
ax.set_title('Radial Profile')
plt.show()
```

```python
r = np.linspace(0.001, 0.5, 100)  # avoid divide by zero at r=0
density = S.mean_radial_profile(r, t=1) / (2 * np.pi * r)
plt.plot(r, density)
plt.xlabel('r')
plt.ylabel('dE/dA')
plt.title('Mean Radial Energy Density')
```
